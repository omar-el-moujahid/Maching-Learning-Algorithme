```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
df=pd.concat([pd.read_csv('./Youtube01.csv'),pd.read_csv('./Youtube02.csv'),pd.read_csv('./Youtube03.csv'),pd.read_csv('./Youtube04.csv'),pd.read_csv('./Youtube02.csv')],ignore_index=True)
```

```python
df=df.drop(columns=['COMMENT_ID','AUTHOR','DATE'],axis=1)
```

```python
df['CLASS'].value_counts()
```

```python
input = df['CONTENT'].copy()
target = df['CLASS'].copy()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_tesy , ytrain , y_test = train_test_split(input,target,test_size=0.2,random_state=365
                                                     ,stratify=target)
```

```python
ytrain.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
ytrain.value_counts(normalize=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
CountVectorizer=CountVectorizer()
x_train_transform = CountVectorizer.fit_transform(x_train)
x_test_transform = CountVectorizer.transform(x_tesy)
```

```python
x_train_transform.toarray()
```

```python
x_test_transform.toarray()
```

```python
x_train_transform.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(x_train_transform,ytrain)
```

```python
clf.get_params()
```

```python
y_preduction = clf.predict(x_test_transform)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(
    y_test, y_preduction , 
    labels=clf.classes_ , cmap='magma'
)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_preduction,target_names=['HAm','Spam']))
```

```python
clf_with_priority=MultinomialNB(class_prior=np.array([0.6,0.4]))
```

```python
clf_with_priority.fit(x_train_transform,ytrain)
```

```python
y_preductin_with_priority=clf_with_priority.predict(x_test_transform)
```

```python
ConfusionMatrixDisplay.from_predictions(
    y_test,y_preductin_with_priority 
    , cmap='magma'
)
```

```python
print(classification_report(y_test,y_preductin_with_priority,target_names=['HAm','Spam']))
```
