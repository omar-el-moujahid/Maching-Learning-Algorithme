```python
df_for_modeling=df[['od_deffrent_time','od_start_time','od_end_time','trip_creation_time','osrm_time route_type_enc','od_start_time osrm_distance','od_end_time osrm_distance','osrm_distance^2','route_type_enc']]
```

```python
from sklearn.model_selection import train_test_split
target=df['start_scan_to_end_scan']
x_train , x_test , y_train , y_test = train_test_split(df_for_modeling,target , test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler 
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.transform(x_test)
```

```python
from sklearn.svm import SVR
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
```

```python
x_train.size
```

```python
svr_model.fit(x_train, y_train)
```

```python
y_pred_cvm=svr_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error 
import numpy as np
import math 
print("Linear Regression Model RMSE is: ", math.sqrt(mean_squared_error(y_test, y_pred_cvm)))
print("Linear Regression Model Training Score: ",svr_model.score(x_train, y_train))
print("Linear Regression Model Testing Score: ",svr_model.score(x_test, y_test))
```
